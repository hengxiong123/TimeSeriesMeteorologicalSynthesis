```python
# TI Curve
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Vhub = np.arange(1,21,1)
Vhub = pd.DataFrame(Vhub)
Iref = [0.18, 0.16, 0.14, 0.12]
Iref = pd.DataFrame(Iref)
Iref.index = ['A+', 'A', 'B', 'C']
b = 5.6
TI= (0.75+b/Vhub).T

# Convert to numpy arrays and perform the outer product to get a 2D result
TI_IEC_ref = pd.DataFrame(Iref.values @ TI.values, columns=TI.columns+1, index=Iref.index) 

TI_IEC_ref
```

```python
TI_IEC_ref.loc['A']
plt.plot(TI_IEC_ref.columns, TI_IEC_ref.loc['A+'])
plt.plot(TI_IEC_ref.columns, TI_IEC_ref.loc['A'])
plt.plot(TI_IEC_ref.columns, TI_IEC_ref.loc['B'])
plt.plot(TI_IEC_ref.columns, TI_IEC_ref.loc['C'])

IEC_string = 'IEC_'
IEC_TI_CLASS = ["A+", "A", "B", "C"]
# IEC_TI_CLASS = [IEC_string + element for element in IEC_TI_CLASS]
plt.xticks(np.arange(1,21,1))
plt.yticks(np.arange(0,1.2,0.1))

plt.legend([IEC_string + element for element in IEC_TI_CLASS])

plt.savefig('IEC_TI_CLASS_Curve.jpg', dpi=600)
```
