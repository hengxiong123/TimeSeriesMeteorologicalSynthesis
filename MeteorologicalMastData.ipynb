```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 文件路径
excel_file_path = r'E:/2023/2023产品开发/TF-SimFARM产品进展汇报_20241022/气象数据/2024.xlsx'

# 要读取的工作表名称
sheet_name_01 = '时间序列'

# 使用pandas的read_excel函数读取指定工作表的数据
df = pd.read_excel(excel_file_path, sheet_name=sheet_name_01)

# 定义风向区间边界及标签，考虑特殊的0代表的区间
# 先正常定义从0到360 + 22.5的单调递增区间，步长为22.5
dir_bin_edges = np.arange(0, 360 + 22.5, 22.5)
dir_bin_labels = [f"{i}-{i + 22.5}" for i in dir_bin_edges[:-1]]

# 处理特殊区间，将348.75 - 11.25对应的区间边界设置正确
# 找到值为360的索引位置，将其改为0
idx_360 = np.where(dir_bin_edges == 360)[0][0]
dir_bin_edges[idx_360] = 0
# 找到值为0的索引位置，将其改为348.75
idx_0 = np.where(dir_bin_edges == 0)[0][0]
dir_bin_edges[idx_0] = 348.75

# 定义风速区间
speed_bin_edges = np.arange(0.5, 20.5 + 1, 1)
speed_bin_labels = [f"{i}-{i + 1}" for i in speed_bin_edges[:-1]]

# 对风向数据进行分组
df['Dir_group'] = pd.cut(df['Dir'], bins=dir_bin_edges, labels=dir_bin_labels, right=False)

# 对风速数据进行分组
df['Speed_group'] = pd.cut(df['Speed'], bins=speed_bin_edges, labels=speed_bin_labels, right=False)

# 先按风向分组，再在每个风向分组内按风速分组，并计算每个分组的大小（即风频）
wind_frequency = df.groupby(['Dir_group', 'Speed_group']).size()

# 计算总数据量
total_count = len(df)

# 计算每个分组的占比
wind_percentage = (wind_frequency / total_count) * 100

# 将wind_frequency转换为DataFrame并设置列名
wind_frequency_df = wind_frequency.reset_index()
wind_frequency_df.columns = ['风向区间', '风速区间', '风频']

# 将wind_percentage转换为DataFrame并设置列名，同时添加百分比符号
wind_percentage_df = (wind_percentage.reset_index() * 100).round(2).astype(str) + '%'
wind_percentage_df.columns = ['风向区间', '风速区间', '占比']

# 打印格式化后的结果
print("风频统计：")
print(wind_frequency_df)
print("占比统计：")
print(wind_percentage_df)

# 设置中文字体为黑体，解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 提取风向区间和风速区间作为x轴标签
x_labels = [f"{wind}_{speed}" for wind, speed in zip(wind_frequency_df['风向区间'], wind_frequency_df['风速区间'])]

# 绘制柱状图
plt.bar(x_labels, wind_frequency_df['风频'])
plt.xticks(rotation=45)
plt.xlabel('风向区间 - 风速区间')
plt.ylabel('风频')
plt.title('不同风向和风速区间的风频分布')
plt.show()
```
